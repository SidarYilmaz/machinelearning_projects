```python
from ultralytics import YOLO

# Hazır eğitilmiş YOLOv8 Nano modelini yüklüyoruz. ('n' nano demek, en hızlı olanıdır)
model = YOLO('yolov8n.pt')

# Modeli indirilen resim üzerinde çalıştırıyoruz.
results = model('test_image.jpg')

# Sonuçları, üzerine kutular çizilmiş yeni bir resim olarak kaydediyoruz.
# Bu işlem 'runs/detect/predict/' klasörünün içine sonuçları kaydeder.
results[0].save(filename='result_image.jpg')

print("✅ Resim analiz edildi ve sonuç 'result_image.jpg' olarak kaydedildi.")
```
